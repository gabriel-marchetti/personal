```python
mnist = tf.keras.datasets.mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
def plot_3x3_image(images, labels):
    plt.figure(figsize=(6, 6))
    for i in range(9):
        plt.subplot(3, 3, i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(images[i], cmap=plt.cm.binary)
        plt.xlabel(f"Label: {labels[i]}", fontsize=7)
        plt.title(f"Image: {i+1}", fontsize=10)
    plt.tight_layout()
    plt.show()
```

```python
plot_3x3_image(x_train, y_train)
```
