```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
college = pd.read_csv('College.csv')
```

```python
college2 = pd.read_csv('College.csv', index_col=0)
college3 = college.rename({'Unnamed: 0' : 'College'}, axis=1)
college3 = college3.set_index('College')
```

```python
college = college3
```

```python
college.describe()
```

```python
interestingAttibutes = college[['Top10perc', 'Apps', 'Enroll']]
pd.plotting.scatter_matrix(interestingAttibutes)
```

```python
idx_notPrivate = college['Private'] == 'No'
idx_Private = college['Private'] == 'Yes'

plt.figure(figsize=(8, 6))
college[idx_notPrivate]['Outstate'].plot(kind='box', color='blue', positions=[1], label='Not Private')
college[idx_Private]['Outstate'].plot(kind='box', color='red', positions=[2], label='Private')

plt.xticks(ticks=[1, 2], labels=['Not Private', 'Private'])
plt.show()
```

```python
college['Elite'] = pd.cut( college['Top10perc'], 
                          [0, 50, 100],
                          labels=['No', 'Yes'] )
college['Elite']
```
